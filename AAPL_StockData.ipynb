```python
from scipy.spatial.distance import euclidean
from fastdtw import fastdtw
```

```python
def dtw_dist(x, y):
    distance, path = fastdtw(x, y, dist=euclidean)
    return distance
```

```python
tseries = []
tlen = 2
for i in range(tlen, len(aapl), tlen):
    pctc = aapl['Close'].iloc[i-tlen:i].pct_change()[1:].values * 100
    res = aapl['Close'].iloc[i-tlen:i+1].pct_change()[-1] * 100
    tseries.append((pctc, res))
```

```python
dist_pairs = []
for i in range(len(tseries)):
    for j in range(len(tseries)):
        dist = dtw_dist(tseries[i][0], tseries[j][0])
        #print(dist)
        dist_pairs.append((i,j,dist,tseries[i][1], tseries[j][1]))
```
